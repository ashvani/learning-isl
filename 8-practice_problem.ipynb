## The Dataset  
- The dataset has 29431 rows and 160 columns
- Six young health participants were asked to perform one set of 10 repetitions of the `Unilateral Dumbbell Biceps Curl` in five different fashions
- They would perform the exercise according to given specification i.e. Class (A, B, C, D and E). This has been recorded in the column `classe` of the dataset and different levels have below meaning  
  1. A -> Correct way to do the exercise
  2. B -> Throwing the elbows to the front
  3. C -> Lifting the dumbbell only halfway
  4. D -> Lowering the dumbbell only halfway
  5. E -> Throwing the hips to the front
- Column `Id` is unique identifier for each row
- 4 sensors are used to record different readings while performing the excercise
- 
  

  

## The Problem
- Treat `classe` variable as target/dependent variable and predict one of the classes out of (A, B, C, D and E) on test data set provided?
- Output should be in the following format having column names `Id` and `pred` and in this order. The column `pred` should have your prediction for that `Id`

  
  

## Approach  
- Read the training data
- Drop the irrelevant columns that will not explain how the excercise was performed such as timestamps, name etc.
- A lot have columns have frequent missing values, how do we handle them?
- A few columns might have missing values but less occuring (e.g. < 5% of the observations are missing for that column). How do we handle these?
- Is this a classification or regression problem?
- Try other feature engineering
- Divide the dataset in train & test set to asses performance of trained model on test set  
- Try different models such as
  1. Logistic Regression
  2. KNN
  3. SVM
  4. Random Forest
- Use Hyperparameter Tuning to select the best set of parameters

- Use the best performing model above to predict on test data set provided in the file `weight_lifting_exercise_test.csv`
