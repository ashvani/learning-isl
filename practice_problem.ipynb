```python
def high_missing_columns(df):
    result = {}
    for column in df.columns:
        count = df[column].isna().sum()
        result[column] = count / df.shape[0]
    return result
```

```python
result = high_missing_columns(df)
```

```python
result = pd.DataFrame(result, index = [0]).T
result = result[result[0] < .7]
result.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
parameters = {'n_estimators': [5, 10, 50, 100, 500],
             'criterion': ['gini', 'entropy'], 
             'max_depth': [10, 100, 500, 1000]}
clf = GridSearchCV(RandomForestClassifier(), parameters, verbose = 5)
clf.fit(X_train, y_train)
```

```python
print(clf.best_score_)
print(clf.best_estimator_)
print(clf.score(X_test, y_test))
```
