## 1 Maximum Margin Classifier   
- $_0 + _1X_1 + _2X_2 = 0$ represents a line (in 2D). In higher dimension it is called a hyperplane.

  
  $_0 + _1X_1 + _2X_2 +    + _pX_p = 0$ .....(1)(defines a p-dimensional plane)

   

- Hyperplanes divide p-dimensional space into 2 halves.
- If y $\in$ {-1, 1} i.e. 2 classes. Suppose if there is a hyperplane that is able to seperate training data perfectly then </br>
  $_0 + _1x_{i1} + _2x_{i2} +    + _px_{ip} > 0$ if $y_i = 1, $  </br>
  and $_0 + _1x_{i1} + _2x_{i2} +    + _px_{ip} < 0$ if $y_i = -1, $ for all i = 1, 2, 3,...., n

- Equivalently,  $y_i(_0 + _1x_{i1} + _2x_{i2} +    + _px_{ip}) > 0$ for all i = 1, 2, 3..., n
- A test observation $x^*$ gets predicted based on sign of $f(x^*) = _0 + _1x^*_{1} + _2x^*_{2} +    + _px^*_{p}$

  


- If the data is seperable by a hyperplane then there will exist infinite number of such hyperplanes. This is because a given separating hyperplane can usually be shifted a tiny bit up or down, or rotated, without coming into contact with any of the observations
- A natural choice is the maximal margin hyperplane (also known as the optimal separating hyperplane), which is the separating hyperplane that is farthest from the training observations

- It is worth misclassifying few observations if we are able to classify other observations really well

- The `support vector classifier` (soft margin classifier) does exactly that

- We allow few observations to be on the incorrect side of the margin or even incorrect side of the hyperplane

- The margin is `soft` because it can be violated

- C is the tuning parameter and M is the width of the margin
- $\epsilon_1$, $\epsilon_2$, ..., $\epsilon_n$ are called slack variables
    1. if $\epsilon_i$ = 0, then ith observation is on the correct side of the margin
    2. if 0 < $\epsilon_i$ <= 1 , then ith observation is on the wrong side of the margin
    3. $\epsilon_i$ > 1, then ith observation is on the wrong side of the hyperplane
- C determines the number and severity of the violations to the margin (and to the hyperplane) that we will tolerate. C is used for bias-variance trade-off

## 3. Support Vector Machines 
- The `support vector machine` (SVM) is an extension of the support vector classifier that results from enlarging the feature space in a specific way, using kernels
- Kernel functions are used to introduce non-linearity
- Examples are:
  1. Linear: <$x, x^{T}$>
  2. Polynomial: $(\gamma<x, x^{T}> + r)^d$
  3. RBF (Radial Basis Function): $e^{-\gamma||x - x^T||^2}$
