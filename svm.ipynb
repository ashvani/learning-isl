## 1 Maximum Margin Classifier   
- $β_0 + β_1X_1 + β_2X_2 = 0$ represents a line (in 2D). In higher dimension it is called a hyperplane.

  
  $β_0 + β_1X_1 + β_2X_2 + · · · + β_pX_p = 0$ .....(1)(defines a p-dimensional plane)

   

- Hyperplanes divide p-dimensional space into 2 halves.
- If y $\in$ {-1, 1} i.e. 2 classes. Suppose if there is a hyperplane that is able to seperate training data perfectly then </br>
  $β_0 + β_1x_{i1} + β_2x_{i2} + · · · + β_px_{ip} > 0$ if $y_i = 1, $  </br>
  and $β_0 + β_1x_{i1} + β_2x_{i2} + · · · + β_px_{ip} < 0$ if $y_i = -1, $ for all i = 1, 2, 3,...., n

- Equivalently,  $y_i(β_0 + β_1x_{i1} + β_2x_{i2} + · · · + β_px_{ip}) > 0$ for all i = 1, 2, 3..., n
- A test observation $x^*$ gets predicted based on sign of $f(x^*) = β_0 + β_1x^*_{1} + β_2x^*_{2} + · · · + β_px^*_{p}$

  


- If the data is seperable by a hyperplane then there will exist infinite number of such hyperplanes. This is because a given separating hyperplane can usually be shifted a tiny bit up or down, or rotated, without coming into contact with any of the observations
- A natural choice is the maximal margin hyperplane (also known as the optimal separating hyperplane), which is the separating hyperplane that is farthest from the training observations
